```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.29,0.76,1.09,1.70,2.,2.52,2.53])
y = np.array([-1.46,-0.98,-0.34,1.78,2.48,5.11,4.92])

t=np.linspace(0,5,400)
quadratic = (t)**2
cubic = (t)**3

plt.plot(t,quadratic, 'g')
plt.plot(t,cubic, 'b')
plt.scatter(x,y,marker='x',c='r')
plt.show()
```
